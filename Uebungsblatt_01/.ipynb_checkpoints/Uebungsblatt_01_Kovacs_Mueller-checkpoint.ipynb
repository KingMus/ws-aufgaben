## 3. Klassifikation
Sie verwenden in dieser Aufgabe das Python-Paket _scikit-learn_.

### 3a) Splitten Sie den Datensatz in Trainings- und Testdaten im Verhältnis 1:9.

```python
x_train, x_test, y_train, y_test = train_test_split(df.drop(["Label", 'clay','sand', "silt", "sampleID", "datetime", "sampleID_hyp"], axis=1), df["Label"],  test_size=0.9)

print("Train", len(x_train))
print("Test", len(x_test))
```

### 3b) Implementieren Sie den Klassifikator Random Forest.  

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=100, oob_score=True, random_state=123456)
rf.fit(x_train, y_train)
predrf = rf.predict(x_test)
```

### 3c) Implementieren Sie den Klassifikator k-Nächste-Nachbarn.

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
knn.fit(x_train, y_train) 
predknn = knn.predict(x_test)
```

### 3d) Implementieren sie ein künstliches Neuronales Netz (ANN) als Klassifikator. 

```python
from sklearn.neural_network import MLPClassifier
ann = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(30, ), random_state=1)
ann.fit(x_train, y_train)
predann = ann.predict(x_test)
```

### 3e) BONUS: Implementieren Sie ein künstliches Neuronales Netz (ANN) in Keras (statt scikit-learn). 

```python
from keras.models import Sequential
from keras.layers import Dense, Activation, Dropout
from keras.utils import to_categorical

from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
encoder.fit(df["Label"])
encoded_y_train = encoder.transform(y_train)
encoded_y_test = encoder.transform(y_test)


model = Sequential()
model.add(Dense(30, input_dim=105, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(10, activation='softmax'))
model.compile(optimizer='adadelta',
              loss='binary_crossentropy',
              metrics=['accuracy'])
model.fit(x_train, to_categorical(encoded_y_train), epochs=20, batch_size=32, verbose=0)
```

### 3f) Tranieren Sie beide Klassifikatoren auf den Trainingsdatensatz und prezidieren Sie auf den Testdatensatz. 

```python
from sklearn.metrics import accuracy_score

score = rf.score(x_test, y_test)
print("Random Forest - Genauigkeit: ", score)

score = knn.score(x_test, y_test)
print("K-Nearest-Neighbour - Genauigkeit: ", score)

score = ann.score(x_test, y_test)
print("ANN (scikit) - Genauigkeit: ",score)

score = model.evaluate(x_test, to_categorical(encoded_y_test))
print("ANN (Keras) - Genauigkeit:", score[1])
```

### 4c) Visualisieren Sie die Klassifikationsergebnisse in geeigneter Form. 

Confusion Matrix --> muss definiert werden. Verwendung einer normalisierten CM (also von 0 bis 1) zur besseren Einschätzung der Genauigkeit.

```python
from sklearn.metrics import confusion_matrix
import itertools

def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    plt.figure(figsize=(20,20))
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

###### Konfusion Martrix Random Forest

```python
cnf_matrix = confusion_matrix(y_test, predrf, labels=y_test.unique() )
plot_confusion_matrix(cnf_matrix , classes=y_test.unique(), normalize=True, title='Normalized confusion matrix of RF')
plt.show()   
```

###### Konfusion Matrix KNN

```python
cnf_matrix = confusion_matrix(y_test, predknn, labels=y_test.unique() )
plot_confusion_matrix(cnf_matrix , classes=y_test.unique(), normalize=True, title='Normalized confusion matrix of KNN')
plt.show()    
```

###### Konfusion Matrix ANN

```python
cnf_matrix = confusion_matrix(y_test, predann, labels=y_test.unique() )
plot_confusion_matrix(cnf_matrix , classes=y_test.unique(), normalize=True, title='Normalized confusion matrix of MLP')
plt.show()   
```
